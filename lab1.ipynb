```python
alpha = 25 - (25 - 1) % 10 - 1 / 10
k = 25 - (25 - 1) % 10 + 1
epsilon = 10**(-5)
```

```python
alpha, k
```

```python
def f(x):
    return (2*(1+alpha)*x**5)+(-3*x**4)+(-1*x**3)+(k*3)
def fp(x):
    return (10*(1+alpha)*x**4)+(-12*x**3)+(-3*x**2)
def sign(x):
    if x > 0:
        return 1.
    elif x < 0:
        return -1.
    elif x == 0:
        return 0.
    else:
        return x   
```

```python
def bisection(a,b,epsilon):
    root = []
    c = (b + a) / 2
    if(f(a) == 0):
        root.append(a)
    if(f(b) == 0):
        root.append(b)
    
    while((b - a) / 2 > epsilon):
        dx = (b - a) / 2
        c = a + dx
        if(sign(f(a))!=sign(f(c))):
            b = c
        else:
            a = c
        
        print("a[i]=%.6lf b[i]=%.6lf f(a)=%.6lf f(b)=%.6lf"%(a,b,f(a),f(b)))
    root.append(c)
    return root
```

```python
def chord(a,b,epsilon):
  
    while (abs(a - b) > epsilon):
        a = b - (b - a) * f(b) / (f(b) - f(a));
        b = a + (a - b) * f(a) / (f(a) - f(b));
        print("a[i]=%.6lf b[i]=%.6lf f(a)=%.6lf f(b)=%.6lf"%(a,b,f(a),f(b)))
    return b;
```

```python
bisection(-100,100,epsilon)
```

```python
chord(-100,75,epsilon)
```

```python
def nuton(x,epsilon):
    x1  = x - f(x)/fp(x)
    x0 = x
    while(abs(x0-x1) > epsilon): 
        x0 = x1
        x1 = x1 - f(x1)/fp(x1)
        print("x_k+1=%lf   f(x_k+1)=%lf"%(x1,f(x1)))
    return x1
```
